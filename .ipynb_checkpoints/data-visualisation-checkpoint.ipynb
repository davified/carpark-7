```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.patches as patches
pd.options.display.max_columns = 30
plt.ion()

plt.rcParams['figure.figsize'] = (16, 4)

plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('./data/monitoring-data.csv')
```

```python
df.head()
```

```python
df['is_treatment'] = np.where(df['Units'].str.contains('CON'), 0, 1)
```

```python
df['Units']
```

## Data cleaning

```python
temp = pd.to_numeric(df['Temperature'], errors='coerce')
```

```python
df['Temperature'] = temp
```

## Data visualisation

```python
df_E9 = df.loc[df['Units'] == 'E9']
```

```python
df_E9.head()
```

```python
my_axis = df_E9.plot(x='Date.1', y='Algae Coverage', kind='bar', title='E9 Algae Coverage')
# df_E9[['Temperature']].plot(secondary_y = True, ax = my_axis)
```

```python
df_E4 = df.loc[df['Units'] == 'E4 (CON)']
df_E4.plot(x='Date.1', y='Algae Coverage', kind='bar', title='E4(CON) Algae Coverage') 
```

```python
df['Algae Coverage'].value_counts()
```

```python
df.groupby('is_treatment')['Algae Coverage'].describe()
```

```python
fig, ax = plt.subplots()

for title, group in df.groupby('is_treatment'):
    color = 'g' if (group['is_treatment'] > 0).all() else 'r'
    group['Algae Coverage'].plot(legend=True, 
                                 title='Algae Coverage (treatment vs. non-treatment)', 
                                 color=color)

ax.legend(["control plot", "treatment plot"]);
```

```python
east_df = df[(df['Units'].str.contains('E'))]
east_df
```

```python
from matplotlib.patches import Rectangle
```

```python
fig, ax = plt.subplots()
for title, group in east_df.groupby('Units'):
    color = 'r' if (group['Units'].str.contains('CON')).all() else 'g'
    ax = group.plot(x='Date.1', y = 'Algae Coverage', ax=ax, color=color)

ax.add_patch(Rectangle((8, 0), 2, 200, facecolor="lightblue"))
plt.annotate('monsoon season', xy=(9, 0), xytext=(8, 0))

ax.legend(["E1", "E2", "E3", "E4(CON)", "E5", "E6", "E7(CON)", "E8", "E9", ]);
```
